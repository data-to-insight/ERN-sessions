```python
list_nums = list(range(1, 21))
for number in list_nums:
    if number %2 == 0: # Note: %x = modulus sign
        print(f'{number} is even')  
    else:
        print(f'{number} is odd')
```

```python
new_nums = range(0,21)
evens = []
odds = []
for number in new_nums:
    if (number != 0) & (number %2 == 0):
        evens.append(number)
        print(f'{number} is not equal to zero and is even')

    elif (number != 0) & (number %2 != 0):
        odds.append(number)
        print(f'{number} is not equal to zero and is odd')
        
    else:
        print(f'{number} is zero')
print(f'Even numbers are: {evens}')
print(f'Odd numbers are: {odds}')

```
