```python
from sqlalchemy import (create_engine, 
                        inspect, 
                        text, 
                        select, 
                        MetaData, 
                        Table, 
                        and_,
                        or_,
                        desc,
                        asc,
                        func,
                        )

engine_903 = create_engine("sqlite+pysqlite:////workspaces/ERN-sessions/intermediate_cohort1_thursday/data/903_database.db")

connection_903 = engine_903.connect()
inspection_903 = inspect(connection_903)

inspection_903.get_table_names()
```

```python
gravity_engine = create_engine("sqlite+pysqlite:////workspaces/ERN-sessions/intermediate_cohort1_thursday/data/gravity.db")
gravity_connection = gravity_engine.connect()
gravity_inspection = inspect(gravity_engine)
gravity_inspection.get_table_names()
```

```python
metadata_903 = MetaData()

episodes = Table('episodes', metadata_903, autoload_with=engine_903)
print(episodes.name)
print(episodes.c.keys())
print(repr(episodes))
```

```python
metadata_gravity = MetaData()

books = Table('book', metadata_gravity, autoload_with=gravity_engine)
print(books.name)
print(books.c.keys())
print(repr(books))
```

```python
with engine_903.connect() as con:
    stmt = "SELECT * FROM episodes"
    result_proxy = con.execute(text(stmt))
    results = result_proxy.fetchall()

first_row = results[0]
print(first_row)
print(first_row.CHILD)
```

```python
header = Table('header', metadata_903, autoload_with=engine_903)

with engine_903.connect() as con:
    stmt = select(header).where(or_(header.columns.ETHNIC == "REFU",
                                    header.columns.ETHNIC == "NOBT"))
    # print(stmt)
    result_proxy = con.execute(stmt)
    results = result_proxy.fetchmany(size=10)

print(results)
```

```python
# select all of the book table where the publication_date is either 1999 or 2000

with gravity_engine.connect() as con:
    stmt = select(books).where(or_(books.columns.book_id == 1,
                                   books.columns.book_id == 2))
    
    result_proxy = con.execute(stmt)
    results = result_proxy.fetchmany(size=10)

print(results)
```
