```python
from sqlalchemy import (create_engine, 
                        inspect, 
                        text, 
                        select, 
                        MetaData, 
                        Table, 
                        and_,
                        or_,
                        desc,
                        asc,
                        func,
                        )

# import pandas as pd

engine_903 = create_engine("sqlite+pysqlite:////workspaces/ERN-sessions/intermediate_friday/data/903_database.db")

connection = engine_903.connect()
inspection = inspect(engine_903)
inspection.get_table_names()
```

```python
engine_chinook = create_engine("sqlite+pysqlite:////workspaces/ERN-sessions/intermediate_friday/data/chinook.db")

connection = engine_chinook.connect()
inspection = inspect(engine_chinook)
inspection.get_table_names()
```

```python
metadata_903 = MetaData()

episodes = Table('episodes', metadata_903, autoload_with=engine_903)

print(episodes.name)
print(episodes.c.keys())
print(repr(episodes))
```

```python
metadata_chinook = MetaData()

albums = Table('albums', metadata_chinook, autoload_with=engine_chinook)

print(albums.name)
print(albums.c.keys())
print(repr(albums))
```

```python
with engine_903.connect() as con:
    stmt = "SELECT * FROM episodes"
    result_proxy = con.execute(text(stmt))
    results = result_proxy.fetchall()

first_row = results[0]
print(first_row)
print(first_row.CHILD)
```

```python
with engine_chinook.connect() as con:
    stmt = "SELECT * FROM albums"
    result_proxy = con.execute(text(stmt))
    results = result_proxy.fetchall()

print(results)
```

```python
import pandas as pd

header = Table('header', metadata_903, autoload_with=engine_903)

# REFU NOBT

with engine_903.connect() as con:
    stmt = select(header).where(or_(header.columns.ETHNIC == "REFU", 
                                    header.columns.ETHNIC == "NOBT"))

    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=20)

df_903 = pd.DataFrame(result)

df_903

# select all of the 'albums' table where the ArtistId is 1 or 2
```

```python
with engine_chinook.connect() as con:
    stmt = select(albums).where(or_(albums.columns.ArtistId == 1,
                                    albums.columns.ArtistId == 2))
    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=10)

albums_df = pd.DataFrame(result)
albums_df
```
