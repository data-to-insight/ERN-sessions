```python
# pip install statsmodels

import pandas as pd
import plotly.express as px
import statsmodels

filepath = '/workspaces/ERN-sessions/data/1980 2023 average house prices.csv'

df = pd.read_csv(filepath)

df.head()
```

```python
# Convert period to date, using the format %Y-%m
df['Period'] = pd.to_datetime(df['Period'], format="%Y-%m", errors="coerce")
```

```python
scatter_1 = px.scatter(df, y='Average price All property types', x='Period')
scatter_1.show()
```

```python
'''In statistics, ordinary least squares (OLS) is a type of linear least squares method for 
choosing the unknown parameters in a linear regression model (with fixed level-one effects of 
a linear function of a set of explanatory variables) by the principle of least squares: 
minimizing the sum of the squares of the differences between the observed dependent variable 
(values of the variable being observed) in the input dataset and the output of the (linear) 
function of the independent variable.'''

scatter_2 = px.scatter(df, 
                       y='Average price All property types', 
                       x='Period',
                       title='Change in Average Property Price for all House Types',
                       labels={'Period':'Time Period',
                               'Average price All property types':'Average house price'},
                       trendline='ols')
scatter_2.show()
```

```python
# I tend to use a categorical histogram rather than a bar chart for this type of data as it removes a step where I need to use a pandas groupby count, and does it for me.
# If you pass plotly a categorical column, it knows what to do!

filepath = r'https://raw.githubusercontent.com/data-to-insight/D2I-Jupyter-Notebook-Tools/main/ml-data%20science%20tutorials/data/CINdetails.csv'

CINdetails = pd.read_csv(filepath)

bar = px.histogram(CINdetails, x='PrimaryNeedCode')

bar.show()
```
