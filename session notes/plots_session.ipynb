```python
import pandas as pd
import plotly.express as px
import statsmodels as sm

filepath = r'https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv'

df = pd.read_csv(filepath)

# df.head()

df['Period'] = pd.to_datetime(df['Period'], format='%Y-%m', errors='coerce')

# df.head()

# matplotlib, seaborn

scatter_1 = px.scatter(df,
                       x='Period',
                       y='Average price All property types')

scatter_1.show()
```

```python
scatter_2 = px.scatter(df,
                       y='Average price All property types',
                       x='Period',
                       title='Average house prices over time',
                       labels={'Period':'Time period',
                               'Average price All property types':'Price (£)'},
                       trendline='ols')
scatter_2.show()
```

```python
filepath = r'https://raw.githubusercontent.com/data-to-insight/D2I-Jupyter-Notebook-Tools/main/ml-data%20science%20tutorials/data/CINdetails.csv'

cin_details = pd.read_csv(filepath)

cin_details['CINreferralDate'] = pd.to_datetime(cin_details['CINreferralDate'], format="%d/%m/%Y", errors='coerce')
cin_details['CINclosureDate'] = pd.to_datetime(cin_details['CINclosureDate'], format="%d/%m/%Y", errors='coerce')
cin_details['DateOfInitialCPC'] = pd.to_datetime(cin_details['DateOfInitialCPC'], format="%d/%m/%Y", errors='coerce')
# cin_details.info()

bar_1 = px.histogram(cin_details, x='PrimaryNeedCode')
bar_1.show()
```

```python
bar_2 = px.histogram(cin_details, x='PrimaryNeedCode', color='ReferralSource', barmode='group')
bar_2.show()
```

```python
cin_details['CPCTimeliness'] = cin_details['DateOfInitialCPC'] - cin_details['CINreferralDate']

cin_details['CPCTimeliness'] = cin_details['CPCTimeliness']/pd.Timedelta(1, 'd')

timeliness_hist = px.histogram(cin_details,
                               x='CPCTimeliness',
                               nbins=20)
timeliness_hist.show()
```

```python
cin_details['Year'] = cin_details['CINreferralDate'].dt.year

timelines_box = px.box(cin_details,
                       y='CPCTimeliness',
                       x='Year')

timelines_box.show()
```

```python
timelines_box_2 = px.box(cin_details,
                       y='CPCTimeliness',
                       color='ReferralSource')

timelines_box_2.show()
```

```python
marginal_scatter = px.scatter(cin_details,
                              x='CINreferralDate',
                              y='CPCTimeliness',
                              # trendline='ols',
                              marginal_x='violin',
                              marginal_y='box',
                              color='PrimaryNeedCode')

marginal_scatter.show()
```

```python
facet_plot = px.histogram(cin_details,
                          x='ReferralSource',
                          facet_col='PrimaryNeedCode',
                          facet_col_wrap=3)

facet_plot.show()
```

```python
cin_details['ReasonForClosure'] = cin_details['ReasonForClosure'].fillna('Still Open')

parallel_cats = px.parallel_categories(cin_details[['PrimaryNeedCode', 'ReasonForClosure']])

parallel_cats.show()
```

```python
cin_details_with_cpc = cin_details[cin_details['CPCTimeliness'].notna()]

size_scatter = px.scatter(cin_details_with_cpc,
                          x='CINreferralDate',
                          y='PrimaryNeedCode',
                          size='CPCTimeliness',
                          color='ReferralSource')
size_scatter.show()
```
