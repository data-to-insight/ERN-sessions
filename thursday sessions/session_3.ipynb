```python
filepath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv"

ci = pd.read_csv(filepath)

print(ci.info())
print(ci.head())
print(ci.tail())

type(ci)
```

```python
# datetime objects
# format argument for datetimes strfstimes
house_prices["Period_dt"] = pd.to_datetime(house_prices["Period"], format="%Y-%m")

house_prices["Period_dt"].info() # check type has changed

house_prices["Period_dt"]
# 01/01/2025 %d/%m%Y
# 01/01/25 %d/%m/%y

# 30-12-2025 %d-%m-%Y
# 30122025 %d%m%Y

```

```python
ci["PersonBirthDate_dt"] = pd.to_datetime(ci["PersonBirthDate"], format="%Y-%m-%d")
ci["ExpectedPersonBirthDate_dt"] = pd.to_datetime(ci["ExpectedPersonBirthDate"], format="%Y-%m-%d")

ci.info()
```

```python
# sum, mean, median, min, max

house_prices_monthly_change_sum = house_prices["Percentage change (monthly) All property types"].sum()
print(f"The sum of percentage changes in house prices across all months is {house_prices_monthly_change_sum}")

max_price = house_prices["Average price All property types"].max()
min_price = house_prices["Average price All property types"].min()

print(f"Max price: {max_price}, \nMin price: {min_price}")

median_price = house_prices["Average price All property types"].median()
mean_price = house_prices["Average price All property types"].mean()

print(f"Median price: {median_price}, \nMean price: {mean_price}")

condition = house_prices["Average price All property types"] < mean_price

less_than_mean = house_prices[condition]

less_than_mean
# calculate total change in house price over time
```

```python
ci["PersonBirthDate_dt"]

collection_end = pd.to_datetime("31/03/2026", format="%d/%m/%Y")

ci["age_delta"] = collection_end - ci["PersonBirthDate_dt"]

ci["age_delta"] = ci["age_delta"] / pd.Timedelta(365.25, "d")

ci["age_delta"] = ci["age_delta"].astype("int")

ci["age_delta"]
```

```python
print(ci["age_delta"].min())
print(ci["age_delta"].max())
print(ci["age_delta"].median())
print(ci["age_delta"].mean())
print(ci["age_delta"].sum())
```
