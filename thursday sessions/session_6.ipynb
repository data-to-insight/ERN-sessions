```python
import pandas as pd

url = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv"

df = pd.read_csv(url)

df.head()
df.info()
```

```python
df['Period_dt'] = pd.to_datetime(df['Period'], format="%Y-%m")

# print(df.columns)

df.columns = map(str.lower, df.columns)

df.columns
```

```python
# Plotly - ploty graph objects, plotly express

import plotly.express as px

price_over_time = px.scatter(df, x='period', y='average price all property types')

price_over_time.show()
```

```python
#install statsmodels
import statsmodels 

price_over_time = px.scatter(df,
                             x='period_dt',
                             y='average price all property types',
                             title='Change in average property price over time (monthly/)',
                             labels={'period':'Time Period',
                                     'average price all property types':'House Price'},
                             trendline="ols"
                             )

price_over_time.show()
```

```python
filepath = r'https://raw.githubusercontent.com/data-to-insight/D2I-Jupyter-Notebook-Tools/main/ml-data%20science%20tutorials/data/CINdetails.csv'

cin = pd.read_csv(filepath)

# cin = cin[['LAchildID', 'CINdetailsID'...]]

cin.drop('Unnamed: 0', axis=1, inplace=True)

cin.info()
```

```python
import plotly.graph_objects as go

bar = px.histogram(cin, x='PrimaryNeedCode')
# title and labels args if we wished 

bar.update_xaxes(categoryorder='category ascending')

bar.show()
```

```python
bar_2 = px.histogram(cin, x='PrimaryNeedCode', color='ReferralSource')

bar_2.show()
```

```python
# CINreferralDate 
# DateOfInitialCPC

# CPCTimeliness - difference between referral and ICPC
# It will be similar to calculating age
# 
# convert boths  columns to dates
# take one date from another
# divide the time difference by 1 day
# set that to a new column

cin['CINreferralDate_dt'] = pd.to_datetime(cin['CINreferralDate'], format="%d/%m/%Y", errors='coerce')
cin['DateOfInitialCPC_dt'] = pd.to_datetime(cin['DateOfInitialCPC'], format="%d/%m/%Y", errors='coerce')

# Possibel to do this for only working days including/excluding bank hols
cin['CPCTimeliness'] = cin['DateOfInitialCPC_dt'] - cin['CINreferralDate_dt']

cin['CPCTimeliness'] = cin['CPCTimeliness']/pd.Timedelta(1, 'd')

timeliness_hist = px.histogram(cin,
                               x='CPCTimeliness',
                               title='Time between Referral and ICPC',
                               nbins=20
                               )

timeliness_hist.show()
```

```python
timeliness_box = px.box(cin,
                        y='CPCTimeliness',
                        color='PrimaryNeedCode')

timeliness_box
```

```python
facet_plot = px.histogram(cin,
                          x='ReferralSource',
                          facet_col='PrimaryNeedCode',
                          facet_row='ReferralNFA')

facet_plot.show()
```

```python
cin_no_nans = cin[cin['CPCTimeliness'].notna()]

size_scatter = px.scatter(cin_no_nans,
                          x='CINreferralDate_dt',
                          y='PrimaryNeedCode',
                          size='CPCTimeliness')

size_scatter.show()
```
