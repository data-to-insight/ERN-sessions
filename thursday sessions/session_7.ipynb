```python
evens = list(range(2,13,2))
odds = list(range(1,13,2))

multiply_list = []

for even in evens:
    for odd in odds:
        result = even * odd
        statement = f'{even} times {odd} is {result}'
        multiply_list.append(statement)

print(multiply_list)
```

```python
# print(6 % 5 == 0)
# print(6 % 1 == 0)

# fizzbang
# take all the numbers from 1-100, 
# print them, BUT if the number is divisible by 3, print fizz,
# if it's divisible by 5 print bang
# BUT if it's divisible by 3 & 5 print fizzbang

nums = list(range(1,101))

for num in nums:
    if (num % 3 == 0) & (num % 5 == 0):
        print('fizzbang') 
    elif num % 3 == 0:
        print('fizz')
    elif num % 5 == 0: 
        print('bang')
    else:
        print(num)
```
